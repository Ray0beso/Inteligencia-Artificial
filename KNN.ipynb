```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.inspection import DecisionBoundaryDisplay

# Cargar el conjunto de datos Wine
wine = datasets.load_wine()

# Tomar las dos primeras caractersticas del conjunto de datos
X = wine.data[:, [0, 1]]
y = wine.target

# Entrenar el clasificador KNN
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

# Graficar la regin de decisin
disp = DecisionBoundaryDisplay.from_estimator(
    knn, X, response_method="predict", cmap=plt.cm.coolwarm, alpha=0.8
)

# Graficar los puntos de datos
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, s=20, edgecolors="k")

# Configurar los ejes y el ttulo del grfico
plt.xlabel("Alcohol")
plt.ylabel("Malic acid")
plt.title("K-Nearest Neighbors Classifier - Wine Dataset")

plt.show()
```
